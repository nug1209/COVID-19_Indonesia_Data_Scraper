```python
import os 

month = "Agustus"

for count, file_name in enumerate(os.listdir("Downloaded/" + month)):
    
    num = int(file_name[2:4])
    new_num = num - 1
    new_num_str = str(new_num)
    print('old name ' + file_name)
    print('new name ' + new_num_str)
    
    if new_num < 10:
        new_num_str = '0' + str(new_num)
    
    old_name = "Downloaded/" + month + "/" + file_name
    new_name = "Downloaded/" + month + "/DL" + new_num_str + file_name[4:]
    
    os.rename(old_name, new_name)


```
